```python
# H0: 구독 여부 간에 컬럼의 평균이 같다.
from scipy.stats import ttest_ind

t_test_results = pd.DataFrame(index=numeric_columns, columns=['t_statistic', 'p_value'])

for column in numeric_columns:
    subscribed = train[train['target'] == 1][column]
    not_subscribed = train[train['target'] == 0][column]
    t_stat, p_value = ttest_ind(subscribed, not_subscribed)
    
    t_test_results.loc[column, 't_statistic'] = t_stat
    t_test_results.loc[column, 'p_value'] = p_value
    
t_test_results
```

T-test 결과, average_time_per_learning_session와 abandoned_learning_sessions 컬럼이 유의수준 5% 하에 통계적으로 유의미한 것으로 나왔다. -> 이 두 컬럼은 구독 여부에 영향이 미칠 가능성이 다소 높다. 추후 변수 중요도를 추가할 예정이다.

예측에 필요한 변수들을 찾기 위해 범주형 변수 간의 독립성 검정을 하는 카이제곱검정을 통해 범주형 변수들이 구독여부와 연관성이 있는지 파악하고자 한다.

```python
## 범주형 변수 간의 독립성 검정 
## H0: 구독여부와 각각의 범주형 변수는 연관성이 없다.

from scipy.stats import chi2_contingency


for column in category_columns:      
    # 교차표
    cont_table = pd.crosstab(train[column], train['target'])

    # 카이제곱 수행
    chi2_stat, p_value, dof, expected = chi2_contingency(cont_table)

    print(column)
    print(f"Chi2 Statistic: {chi2_stat}")
    print(f"P-value: {p_value}")
    print(f"Degrees of Freedom: {dof}")
    print("Expected Frequencies Table:")
    print(expected)
    print('---------------------------------------------------------')
    
```

각 범주형 변수마다 p-value를 살펴보니 유의수준 5% 하에서 모두 귀무가설을 기각할 수 없다는 결론이 나왔다. 이는 구독 여부와 지불패턴, 구독유형, 선호 난이도는 연관성이 없다는 얘기인데 범주형 변수를 예측에 사용해야할지 말지는 좀더 생각해봐야겠다.
