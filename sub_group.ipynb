```python
import pandas as pd
info = pd.read_csv('/content/drive/MyDrive/predict_subcribe/data_info.csv')
sub = pd.read_csv('/content/drive/MyDrive/predict_subcribe/sample_submission.csv')
train = pd.read_csv('/content/drive/MyDrive/predict_subcribe/train.csv')
test = pd.read_csv('/content/drive/MyDrive/predict_subcribe/test.csv')
```

```python
train = train.drop(['user_id'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
labels = [col for col in train.columns if train[col].dtype == 'object' and col not in ['user_id']]
label_encoders = {}
for col in labels:
    label_encoders[col] = LabelEncoder()
    train[col] = label_encoders[col].fit_transform(train[col])
    test[col] = label_encoders[col].transform(test[col])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
k=5
X = train
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
```

```python
train['cluster'] = kmeans.labels_
train['group'] = kmeans.labels_
```

```python
group_summary1 = train.groupby('group')['subscription_duration','recent_login_time',].describe()
group_summary2 = train.groupby('group')['average_login_time','average_time_per_learning_session',].describe()
group_summary3 = train.groupby('group')['monthly_active_learning_days','total_completed_courses',].describe()
group_summary4 = train.groupby('group')['recent_learning_achievement','abandoned_learning_sessions',].describe()
group_summary5 = train.groupby('group')['community_engagement_level','preferred_difficulty_level',].describe()
group_summary6 = train.groupby('group')['subscription_type','customer_inquiry_history',].describe()
group_summary7 = train.groupby('group')['payment_pattern','target',].describe()


display(group_summary1)
display(group_summary2)
display(group_summary3)
display(group_summary4)
display(group_summary5)
display(group_summary6)
display(group_summary7)
```

```python
train['harder'] = 0
train.loc[(train['average_time_per_learning_session'] > 144), 'harder'] = 1
test['harder'] = 0
test.loc[(test['average_time_per_learning_session'] > 144), 'harder'] = 1
```

```python
train['degree'] = train['monthly_active_learning_days'] * train['average_time_per_learning_session']
train['concentrate'] = train['average_time_per_learning_session'] - train['average_login_time']

test['degree'] = test['monthly_active_learning_days'] * test['average_time_per_learning_session']
test['concentrate'] = test['average_time_per_learning_session'] - test['average_login_time']
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
train = TabularDataset(train)
test = TabularDataset(test)
```

```python
fit_parameters = {
    'presets': 'best_quality',
    'num_bag_folds': 7,
    'num_bag_sets': 5,
    'num_stack_levels': 1

} #auto_stack = True, fit_weighted_ensemble = True,
```

```python
predictor = TabularPredictor(label='target', eval_metric='f1_macro',sample_weight='balance_weight')
predictor.fit(train, auto_stack = True, fit_weighted_ensemble = True,presets=fit_parameters, num_gpus=1,hyperparameter_tune_kwargs='auto')
```

```python
predictor.feature_importance(data=train)
```

```python
pred_y = predictor.predict(test)
pred_y
```

```python
sub['target'] = pred_y
sub.to_csv('auto_new_col3_87.csv',index=False)
```
